```python
import pandas as pd
import numpy as np
from better_profanity import profanity 
```

```python
df = pd.read_csv('/content/dataset.csv')
```

```python
df1 = df.drop(['count', 'hate_speech', 'offensive_language', 'neither', 'class'], axis = 1)
```

```python
df1 = df1.iloc[:, 1:]
```

```python
df1
```

```python
np.savetxt("test_file.txt", df1, fmt = "%s")
```

```python
titles = []

with open("/content/test_file.txt", "rt", encoding = "utf-8") as f:
    lines = f.readlines()
    for x in lines:
        line = x.strip("\n")
        titles.append(line)
```

```python
cuss_words = []

for x in titles:
    c = 0
    for i in x.split():
        value = profanity.contains_profanity(i)
        if value == True:
            c += 1
    cuss_words.append(c)
```

```python
data = {"Sentences":titles,
        "Curse Words":cuss_words}

df2 = pd.DataFrame(data)
```
